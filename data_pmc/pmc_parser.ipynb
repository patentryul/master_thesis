```python
# import
from xml.etree import ElementTree as ET
from xml.etree.ElementTree import Element, dump, ElementTree

# Use Open source 'pubmed_parser'
import pubmed_parser as pp
```

```python
import glob

targetPattern = r"pmc_xml/*.xml"
xmlList = glob.glob(targetPattern)

for i in range(len(xmlList)):
    xmlList[i] = xmlList[i].replace('\\', '/')
    print(xmlList[i])
```

```python
drug_doc = {} # dictionary
for xml in xmlList:
    tree = ET.parse(xml)
    root = tree.getroot()
    dict_out = pp.parse_pubmed_xml(xml)
    fileName = xml.split('/')[1]
    fileName = fileName.split('.')[0]
    drug_doc[fileName] = dict_out
```
