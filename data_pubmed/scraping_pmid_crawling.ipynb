```python
import requests
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait

url = 'https://pubmed.ncbi.nlm.nih.gov/'
```

```python
# Drug bank common name list
import pandas as pd

df = pd.read_csv('D:/WorkSpace/master_thesis/Drug_Bank/drugbank_vocabulary.csv')
drugName = df['Common name']
```

```python
def save_pmid(drugName): 
    # Drug name으로 검색     
    search_window  = driver.find_element_by_id('id_term')
    search_window.send_keys(Keys.CONTROL + "a", Keys.DELETE) # 검색 전 검색창 초기화
#     driver.implicitly_wait(1)
    search_window.send_keys(drugName)
    search_window.submit()
#     search_window.send_keys(Keys.ENTER)       
        
    # Save file - Click button
    save_btt  = driver.find_element_by_xpath('//*[@id="save-results-panel-trigger"]')
    save_btt.click()

    # Save file - select options
    select = Select(driver.find_element_by_id('save-action-selection'))
    select.select_by_value('all-results')
    
    select = Select(driver.find_element_by_id('save-action-format'))
    select.select_by_value('pmid')
    
    # Create file
    create_btt  = driver.find_element_by_xpath('//*[@id="save-action-panel-form"]/div[3]/button[1]')
    create_btt.click()
```

```python
def retry_connection(drugName):
    driver.quit()
    driver = webdriver.Chrome(executable_path=r'D:\WorkSpace\master_thesis\chromedriver_win32\chromedriver.exe', options=chrome_options)
    driver.get(url)
    save_pmid(drugName)
```
