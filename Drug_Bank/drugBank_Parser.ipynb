```python
# nan 여부 확인 함수
def isNaN(num):
    return num != num
```

```python
# Name 정리 (Common & Synonyms)
drug_list = list()
for i in range(len(df)):
    common = df['Common name'][i]
    synonyms = df['Synonyms'][i]    
    if isNaN(synonyms):
        drug_list.append(common)
    else:
        drug_list.append(common + ' | ' + synonyms)
            

print(len(drug_list))
```
